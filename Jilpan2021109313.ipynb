```python
# Define response messages
response_messages = {
    "Hello": "Hi there!",
    "How are you?": "I'm doing well, thank you for asking!",
    "What's your favorite color?": "I don't have a favorite color, but I can help you with any questions you have.",
    "Tell me a joke": "Why don't scientists trust atoms? Because they make up everything!",
    "Can you help me with programming?": "Sure, I can assist you with programming questions. What specifically do you need help with?"
}

# Welcome message
print("Welcome to the AI Chatbot. How can I assist you today?")

# Chat loop
while True:
    # User input
    user_input = input("You: ")

    # Exit condition
    if user_input.lower() == 'exit':
        print("ChatBot: Goodbye!")
        break

    # Check if user input exists in response messages
    if user_input in response_messages:
        # If yes, print corresponding response
        print("ChatBot:", response_messages[user_input])
    else:
        # If no, print a generic message
        print("ChatBot: Sorry, I don't understand that.")
```
